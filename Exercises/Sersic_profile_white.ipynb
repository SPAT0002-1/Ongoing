# Simulation of 1D galaxy luminosity profile 

The azimuthally averaged luminosity profile of a galaxy can be expressed as a Sersic profile of index $n$ with the following expression: 

$$
I = I_e * \exp\left(-b * \left(\left(\frac{\theta}{\theta_e}\right)^{1./n} - 1.\right) \right)
$$

where $b$ can be approximated as $b = 2*n - 1./3$, $n$ is the Sersic index, and $\theta_e$ is the angular effective radius. The two cases $n=1$ correspond to the exponential profile characteristic of a spiral galaxy, and $n=4$ to a de Vaucouleurs profile, characteristic of the properties of an elliptical galaxy. 
For a specific analysis, you need to simulate a profile with $n=4$, $\theta_e = 50 $ pixels, and $I_e = 100$ e/s (electrons per second), on a grid of 100 pixels. Note that you evaluate the profile in the center of each pixel (i.e. for the first pixel, you will calculate I at $\theta = 0.5$ pixels). The question you aim to answer for this simulated profile are: 
(i) what is the uncertainty $\sigma_{\theta_e}$ on $\theta_e$ as a function of the noise $\sigma$ on the pixel intensity ?; (ii) How much is the uncertainty affected by the knowledge of the Sersic index $n$ ?   
To answer these questions, you will follow the procedure outlined below: 

* (1) Create a function that returns a (one dimensional) Sersic profile. Write your function such that the `help` command gives a description of your function.   
* (2) Create a second function that returns a Sersic profile for which a noise (assumed to be a gaussian with width $\sigma$) is added to each pixel.   
* (3) Generate a noisy realisation (sigma=300) of your sersic profile with $n=4$, $\theta_e = 50 $ pixels, and $I_e = 100$ e/s. Generate a figure (with labels) showing your simulated data as points with error bars, and overplot the model (without noise) as a continuous line.    
* (4) Use a frequentist approach to fit the model defined in (1) to one noisy realisation of your data.    
* (4b) Print at the screen the values of the parameters with only 2 digits.    
* (5) Estimate the uncertainty $\sigma_{\theta_e}$ on $\theta_e$ for different values of the noise $\sigma$ (e.g. sigma=10, 50, 100, 200, 300, 400, 500). Make a figure showing $\sigma_{\theta_e}$ as a function of $\sigma$.        
* (6) Create a function similar to the one you defined in (1) but where the Sersic index $n$ is fixed to $n=4$ (call this function e.g. `devauc()`). Compare the uncertainty on $\theta_e$ with the results found in (5).
