# Atmospheric CO2 at Mauna Loa observatory


Since 1958 the scientists at the observatory of Mauna Loa are recording the local amount of CO2 in the atmosphere on a monthly basis. This observatory is situated at 3400m above sea level on the slope of the volcano Mauna Loa on the island of Hawaii at the position (19° 28' 46,3" N, 155° 36' 09,6" W). Its position far from any pollution source make it a perfect place for atmospheric measurements.

In this exercise you will have access to Mauna Loa's observatory raw data and will have to apply several treatment to it in order to isolate annual variation and global trends.

## Data

The file "monthly_in_situ_co2_mlo.csv" contains monthly measurements taken in Mauna Loa. The time is kept in decimal years and the measurements are done in the middle of each month. Which means that January 1958 is recorded as 1958 + (0.5 /12) =  1958,042, February 1958 as  1958 + (1,5 / 12) = 1958,125 an so on. The first record of the file was done in March 1958 and the last in September 2023. Few lines exist before and after these two dates but the values are fixed to "-99.99" which means the data wasn't available when this file was downloaded.

The CO2 measurements are given in micro-mole per mole of dry air (µmol/mol, or ppmv)
